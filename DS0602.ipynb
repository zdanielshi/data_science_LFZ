# Formula 1 Exercise

Connect to the web server and make sure you receive an 'ok' status code

```python
f1_url = "https://www.formula1.com/en/results.html/2022/drivers.html"
```

```python
f1_page = requests.get(f1_url)
f1_page
```

Scrape the page for the statistics stored in the table

```python
f1_raw_data = BeautifulSoup(f1_page.content)
```

```python
f1_table = f1_raw_data.find_all('table')
f1_table = f1_table[0]
```

```python
f1_table_rows = f1_table.find_all('tr')
```

```python
len(f1_table_rows)
```

```python
f1_rankings = []

for x in f1_table_rows:
  td = x.find_all('td')
  info = [y.text for y in td]
  f1_rankings.append(info)

f1_rankings
```

Save the table as a data frame

```python
df_f1_rankings = pd.DataFrame(f1_rankings)
df_f1_rankings = df_f1_rankings.iloc[1:, 1:6] #there was an empty column that got scraped
df_f1_rankings.columns = ['POS', 'DRIVER', 'NATIONALITY', 'CAR', 'PTS']
df_f1_rankings.head()
```

Replace the \n's in the dataframe

```python
df_f1_rankings.replace({'\n': ' '}, regex = True, inplace = True)
df_f1_rankings.head()
```

Save the data frame as a .csv file, and re-read the file 
to make sure it saved correctly
