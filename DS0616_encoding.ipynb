```python
import pandas as pd
```

```python
titanic = pd.read_csv('/content/titanic.csv')
```

```python
titanic
```

```python
titanic.isna().sum()
```

```python
titanic.drop(columns = ['PassengerId', 'Name', 'Cabin'], inplace = True)
```

```python
titanic.Age.fillna(titanic.Age.mean(), inplace=True)
titanic.Embarked.fillna(titanic.Embarked.mode(), inplace=True)
```

```python
titanic.isna().sum()
```

```python
titanic.SibSp = titanic.SibSp.astype('str')
titanic.Parch = titanic.Parch.astype('str')
```

```python
pd.get_dummies(titanic)
```

```python
import category_encoders as ce
```

```python
# do this for the categorical features, that don't have ordinal values
one_hot = ce.OneHotEncoder(cols=['Sex', 'Parch', 'SibSp', 'Ticket', 'Embarked'])
```

```python
encoded = one_hot.fit_transform(titanic)
```

```python
encoded
```

```python
mapping = [{
    "col": "Pclass",
    "mapping": [
                ('1','3'),
                ('2','2'),
                ('3','1')
    ]
}]
```

```python
ordinal = ce.OrdinalEncoder(mapping = mapping, cols = ['Pclass'])
```

```python
encoded = ordinal.fit_transform(titanic)
```
