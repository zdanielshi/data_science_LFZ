Use groupby some groups, to get more granular averages

```python
titanic.groupby(['Sex','Pclass'])['Age'].mean().plot(kind='bar')
```

```python
titanic.Age.fillna(titanic.groupby(['Sex','Pclass'])['Age'].transform('mean'),inplace=True)
```
