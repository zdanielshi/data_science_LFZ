```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
titanic = pd.read_csv('/content/titanic.csv')
titanic.head()
```

```python
titanic.info()
```

```python
plt.plot(titanic['Age'])
```

```python
plt.hist(titanic.Age, color = 'purple', bins = 50)
```

```python
titanic.hist(figsize=(10,6));
plt.tight_layout
```

```python
plt.boxplot(titanic.Age.dropna());
```

Build 2 histograms
* For males
* For females
* See if the distributions are similar

```python
males = titanic[titanic.Sex == 'male']
females = titanic[titanic.Sex == 'female']
```

```python
plt.figure(figsize = (10,6))
plt.hist(males.Age, color = 'blue', bins = 20, alpha = 0.5, label = 'Male');
plt.hist(females.Age, color = 'pink', bins = 20, alpha = 0.5, label = 'Female');
plt.title('Age Distribution of Titanic Data')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.savefig('Age_Distribution_image.png')
```

```python
plt.scatter(titanic.Age, titanic.Fare)
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
plt.scatter(titanic.Pclass, titanic.Fare)
plt.xlabel('Pclass')
plt.ylabel('Fare')
```

Do it with a aggregate function of mean and sum

```python
titanic.groupby(['Sex', 'Pclass'])['Survived'].agg(['sum','mean'])
```

Scatter of Fare and Age for people who survived vs not survived. 

```python
survived = titanic[titanic.Survived == 1]
unalived = titanic[titanic.Survived == 0] 
```

```python
plt.figure(figsize = (15,10))
plt.scatter(survived.Fare, survived.Age, alpha = 0.25, color = 'royalblue', label = 'Survived')
plt.scatter(unalived.Fare, unalived.Age, alpha = 0.25, color = 'salmon', label = 'Unalived')
plt.title('Distribution of Age and Fare for Survived vs Unalived Passengers')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.legend()
plt.savefig('Age v Fare for Survived v Not Survived.png')
```
