Build 2 histograms
* For males
* For females
* See if the distributions are similar

```python
males = titanic[titanic.Sex == 'male']
females = titanic[titanic.Sex == 'female']
```

```python
plt.figure(figsize = (10,6))
plt.hist(males.Age, color = 'blue', bins = 20, alpha = 0.5, label = 'Male');
plt.hist(females.Age, color = 'pink', bins = 20, alpha = 0.5, label = 'Female');
plt.title('Age Distribution of Titanic Data')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.savefig('Age_Distribution_image.png')
```

```python
plt.scatter(titanic.Age, titanic.Fare)
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
plt.scatter(titanic.Pclass, titanic.Fare)
plt.xlabel('Pclass')
plt.ylabel('Fare')
```

Do it with a aggregate function of mean and sum

```python
titanic.groupby(['Sex', 'Pclass'])['Survived'].agg(['sum','count'])
```

Scatter of Fare and Age for people who survived vs not survived. 

```python
survived = titanic[titanic.Survived == 1]
unalived = titanic[titanic.Survived == 0] 
```

```python
plt.figure(figsize = (15,10))
plt.scatter(survived.Age, survived.Fare, alpha = 0.25, color = 'royalblue', label = 'Survived')
plt.scatter(unalived.Age, unalived.Fare, alpha = 0.25, color = 'salmon', label = 'Unalived')
plt.title('Distribution of Age and Fare for Survived vs Unalived Passengers')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.legend()
plt.savefig('Age v Fare for Survived v Not Survived.png')
```

# Healthcare Data Viz

## Loading the data

```python
stroke = pd.read_csv('/content/healthcare-dataset-stroke-data (1).csv')
stroke
```

```python
stroke.info()
```

## Cleaning up the data

```python
stroke.stroke.value_counts(normalize=True)
```

```python
stroke.ever_married.value_counts(normalize=True)
```

```python
stroke.Residence_type.value_counts(normalize=True)
```

```python
# COME BACK TO THIS ONE AND TRY TO DO A PIE CHART
```

```python
had_stroke = stroke[stroke.stroke == 1]
no_stroke = stroke[stroke.stroke == 0]
```

```python
reorder = stroke.pop('stroke')
stroke.insert(0,'stroke', reorder)
stroke

stroke_corr = stroke.corr().sort_values('stroke', ascending=False)
stroke_corr 
```

## Visualizations

```python
had_stroke.hist(figsize = (10,5));
no_stroke.hist(figsize = (10,5));
```

```python
# compare glucose levels
plt.figure(figsize = (10,6)) # fig size
plt.title('Glucose Levels for Had Stroke vs No Stroke') # title
plt.hist(had_stroke.avg_glucose_level, color = 'darkred', bins = 25, alpha = 0.4, label = 'Had Stroke'); # series for stroke
plt.hist(no_stroke.avg_glucose_level, color = 'grey', bins = 25, alpha = 0.4, label = 'No Stroke'); # series for no stroke
plt.legend(); # legend
```

```python
# compare BMI
plt.figure(figsize = (10,6)) # fig size
plt.title('BMI for Had Stroke vs No Stroke') # title
plt.hist(had_stroke.bmi, color = 'darkred', bins = 25, alpha = 0.4, label = 'Had Stroke'); # series for stroke
plt.hist(no_stroke.bmi, color = 'grey', bins = 25, alpha = 0.4, label = 'No Stroke'); # series for no stroke
plt.legend(); # legend
```

```python
# compare age
plt.figure(figsize = (10,6)) # fig size
plt.title('Age for Had Stroke vs No Stroke') # title
plt.hist(had_stroke.age, color = 'darkred', bins = 25, alpha = 0.4, label = 'Had Stroke'); # series for stroke
plt.hist(no_stroke.age, color = 'grey', bins = 25, alpha = 0.4, label = 'No Stroke'); # series for no stroke
plt.legend(); # legend
```


Some scatter comparisons

```python
plt.scatter(no_stroke.age, no_stroke.bmi, color = 'red', alpha = 0.1, label = 'No Stroke');
plt.scatter(had_stroke.age, had_stroke.bmi, color = 'royalblue', alpha = 0.5, label = 'Stroke');
plt.title('Age and BMI for Stroke vs No Stroke')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend();
```

```python
plt.scatter(no_stroke.age, no_stroke.avg_glucose_level, color = 'red', alpha = 0.1, label = 'No Stroke');
plt.scatter(had_stroke.age, had_stroke.avg_glucose_level, color = 'royalblue', alpha = 0.5, label = 'Stroke');
plt.title('Age and Avg Glucose Level for Stroke vs No Stroke')
plt.xlabel('Age')
plt.ylabel('Average Glucose Level')
plt.legend();
```
